```python
import pandas as pd
df=pd.read_csv('/home/bot/Desktop/intern/datasets/sales_prediction.csv')
df1=pd.read_csv('/home/bot/Desktop/intern/datasets/sales_prediction.csv')
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df1.drop(['Unnamed: 0'],axis=1,inplace=True)

print(df)
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i in df:
    plt.plot(df[f'{i}'],df.Sales,'^')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
for i in df:
    df[[f'{i}']]=ss.fit_transform(df[[f'{i}']])
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg
```

```python
reg.fit(df[['TV','Radio','Newspaper']],df.Sales)
```

```python
p=reg.predict(df[['TV','Radio','Newspaper']])
```

```python
from sklearn import metrics
print(metrics.r2_score(df.Sales,p))
```

```python
reg.fit(df1[['TV','Radio','Newspaper']],df1.Sales)
p1=reg.predict(df1[['TV','Radio','Newspaper']])
print(metrics.r2_score(df1.Sales,p1))
```
