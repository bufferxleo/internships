```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/home/bot/Desktop/interships/yhills/datasets/titanic-training-data.csv")
```

### single cateforical -> countplot
### single numerical -> distplot
### two cate -> countplot
### two numerical -> scatterplot
### one categorical, one num -> use other plots strip,violin etc.


```python
plt.figure(figsize=(8,6))
#create fig size 

#create countplot for passengers by class

sns.countplot(x='Pclass',data=df)
plt.title('passengers by class')
plt.xlabel('class')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
#create fig size 

#create countplot for passengers by class

sns.violinplot(x='Pclass',y='Age',data=df)
plt.title('Age distribution by class')
plt.xlabel('class')
plt.ylabel('age')
plt.show()


#violin plot gives info about the inferece blw classes
```
